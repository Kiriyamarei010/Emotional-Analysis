```python
import numpy as np
import matplotlib.pyplot as plt

# Emotion data for Segment 1
Segment1_emotions = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'shame']
Segment1_scores = [17.547, 0.290031, 9.8378, 38.1596, 2.014, 27.5012, 4.7052]

# Emotion data for Segment 2
Segment2_emotions = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'shame']
Segment2_scores = [14.4305, 0.0924022, 12.0211, 40.3105, 0.449532, 31.7521, 0.958083]

Segment3_emotions = ['anger', 'disgust', 'fear', 'joy', 'neutral', 'sadness', 'shame']
Segment3_scores = [7.98417, 0.0708961, 10.5578, 44.1309, 0.197827, 36.4116, 0.627521]



# Function to plot radar chart
def plot_radar_chart(emotions, scores, label, ax):
    angles = np.linspace(0, 2 * np.pi, len(emotions), endpoint=False).tolist()
    scores += scores[:1]
    angles += angles[:1]

    ax.plot(angles, scores, 'o-', linewidth=2, label=label)
    ax.fill(angles, scores, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(emotions)
    ax.grid(True)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Plot radar chart for Segment 1
plot_radar_chart(Segment1_emotions, Segment1_scores, 'Cigarettes', ax)

# Plot radar chart for Segment 2
Segment2_scores = Segment2_scores[:len(Segment2_emotions)]
plot_radar_chart(Segment2_emotions, Segment2_scores, 'ptsd and effects', ax)

Segment3_scores = Segment3_scores[:len(Segment3_emotions)]
plot_radar_chart(Segment3_emotions, Segment3_scores, 'opening up', ax)


# Add a legend
ax.legend(loc='upper right')

# Display the chart
plt.show()
```
